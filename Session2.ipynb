We start the preparation of the data.
We will look for missing values and outliers.
Missing values are easier, we just need the right functions.

Looking for missing values using .isnull and summarizing the number of the missing values throughout the database.
We need sum() twice to add missing values in each row and column.

```python
df.isnull().sum().sum()
```

We found 13 missing values (represented by NaN in the data table).
We want to know where are these missing values.
We ask to print the number of missing values in each column (we add only by columns).

```python
df.isnull().sum()
```

We see that all missing values are in the Area column.
We know from the description of the database that this column is a simple coding of the Address column, so
*   when Address = "Adam Road", Area = 0
*   when Address = "Book Street", Area = 1
*   when Address = "Station Corner", Area = 2






Solving the missing value problem is simple, we can put 1 as all missing values are on Book Street.

```python
df['Area'].fillna(value=1, inplace = True)
```

We used the 'Area' column (variable), we applied the fillna method (yes, filling NAs) with the value of 1, and we used the inplace argument to keep the changes in df.

We can cross-check if everything is alright.

```python
df.isnull().sum()
```

Yes, 0 missing values.

Evidently, depending on the nature of missing values, more steps, and more complex filtering may be needed. There is no one-size-fits-all solution for missing values (see the lecture).
